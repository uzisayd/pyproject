```python
import re
from datetime import datetime

def name():
    name = input("Enter name: ")
    pattern = "^[A-z]*[a-z]$"
    if re.match(pattern, name):
        return True
    else:
        print("Invalid name.")
        return False

def dob():
    dob = input("Enter date of birth (DD-MM-YYYY): ")
    pattern = "^[0-9]{2}-[0-9]{2}-[0-9]{4}$"
    if re.match(pattern, dob):
        try:
            dob_date = datetime.strptime(dob, "%d-%m-%Y")
            if dob_date < datetime.now():
                return True
            else:
                print("Date of birth cannot be in the future.")
                return False
        except ValueError:
            print("Invalid date format.")
            return False
    else:
        print("Invalid date format. Use DD-MM-YYYY.")
        return False

def mobile():
    mobile = input("Enter mobile number: ")
    pattern = "^[6-9][0-9]{9}$"
    if re.match(pattern, mobile):
        return True
    else:
        print("Invalid mobile number. Must be exactly 10 digits.")
        return False

def email():
    email = input("Enter email: ")
    pattern = "^[a-z][a-zA-Z0-9._]*@[a-z]*.com$"
    if re.match(pattern, email):
        return True
    else:
        print("Invalid email address.")
        return False

def customer_id():
    customer_id = input("Enter customer ID: ")
    pattern = "^[1-9][0-9]{4}$"
    if re.match(pattern, customer_id):
        return True
    else:
        print("Invalid customer ID.")
        return False

def address():
    address = input("Enter address: ")
    if len(address.strip()) > 0:
        return True
    else:
        print("Address cannot be empty.")
        return False

def gender():
    gender = input("Enter gender (male/female/other): ")
    l1 = ["male", "female", "others"]
    if gender in l1:
        return True
    else:
        print("Invalid gender. Must be 'male', 'female', or 'other'.")

def number_of_people():
    number = input("Enter number of people (Maximum 12): ")
    pattern = "^[1-9][0-2]$"
    if re.match(pattern, number):
        return True
    else:
        print("Invalid number of people.")
        return False

print("\n------------------------------------------------------")

print("\nValidating customer details...")

customer_valid = True

if not name():
    customer_valid = False
if not dob():
    customer_valid = False
if not mobile():
    customer_valid = False
if not email():
    customer_valid = False
if not customer_id():
    customer_valid = False
if not address():
    customer_valid = False
if not gender():
    customer_valid = False

if customer_valid:
    print("Customer details are valid!")
else:
    print("Customer details are invalid!")

print("\n------------------------------------------------------")
print("\nValidating service details...")
service_valid = True
if not customer_id():
    service_valid = False

if not number_of_people():
    service_valid = False

if service_valid:
    print("Service details are valid!")
else:
    print("Service details are invalid!")

print("\n------------------------------------------------------")
```
